```python
from rembg import remove
from PIL import Image
import io

# Define input and output paths

def process(input_path, output_path, new_size=(64,64)):
    # Open the input image
    with Image.open(input_path) as input_image:
        # Remove the background
        output_image = remove(input_image)
        output_image = output_image.resize(new_size)

        # Save the resulting image with a transparent background
        # Ensure to save as PNG to support transparency
        output_image.save(output_path)

    print(f"Background removed and saved to {output_path}")
```

```python
import os
contents = os.listdir("C:/Users/there/Desktop/DnD/public/characters/original/")

for file in contents:
    nnm = ".".join(file.split(".")[:-1])
    process(f"C:/Users/there/Desktop/DnD/public/characters/original/{file}", f"C:/Users/there/Desktop/DnD/public/characters/processed/{nnm}.png")
```
